### Handling Missing/Null Values

cells. It is good practice to deal with this missing values. Missing values might sometimes lead to false prediction of model. Using pandas you can check if the provided data has a missing values and you can check each column with empty values. There are two different methods to deal with Null data:
- Drop i.e., delete the row/column from the dataset

### Fillna

The empty data from the dataset as to be dealt though some strategic planning. Pandas allows you to fill the empty values using fillna method. But you should use this command precisely, a wrong entered value might cause a major problem in Analysis.
Consider two scenarios, first you have the dataset of Students with their total marks in 5 different subjects. And in second scenario you have the dataset of 8 Companies Sales from May 2020 To July 2021. In both the Datasets there might be empty values.
How will you deal with Empty Values in the above given dataset?

Consider the first dataset, the data is empty if in case the student is absent for the exam. You can then fill the empty values with 0. 
In second dataset, since you are dealing with sales, there might be empty data. This empty data should contain a price for the specific month, but is left empty. In such cases you can take the average mean of that column (here column is Company Name).

#### `bfill` and `ffill` arguments in Fillna

````python
check_ffill=pd.DataFrame({"ColOne":[6,None,None,10,5],
                 "ColTwo":[None,7,2,3,None],
                 "ColThree":[4,None,8,5,None],
                 "ColFour":[5,4,2,None,0]})
````

````python
check_ffill  
````

````python
check_ffill.fillna(method="ffill")  
#forward fill, check the previous row value and update it. Thus ColTwo first data is still NaN
````

````python
check_ffill.fillna(method="bfill")  #backward fill, thus the last data are still NaN
````

### Parsing Dates In Pandas

On working with dates let us take a dataset that has Date column in it. Luckily in [Pandas Part-1](https://github.com/lucifertrj/100DaysOfML/tree/main/Day2:Pandas) we used a dataset with named data.csv. We shall import the data.csv file, make sure you provide file path properly. 

````python
working_with_date = pd.read_csv("data.csv")
````

````python
working_with_date.info()
````

If you check the datatype of date_added it is an object. Basically in almost many dataset date shall be in object data type. We can convert the object to datetime with help of Powerful Pandas

Let us create a new column

````python
working_with_date['date'] = pd.to_datetime(working_with_date['date_added'])
````

````python
working_with_date.info()
````

Since we now have a seperate column for date lets first delete the object date_added column 

````python
working_with_date.drop('date_added',axis=1,inplace=True)
````

````python
working_with_date.date
````

Manipulating Year, Month and Day of datetime column

````python
working_with_date['year'] = working_with_date['date'].dt.year
````

````python
working_with_date['month'] = working_with_date['date'].dt.month
````

````python
working_with_date['day'] = working_with_date['date'].dt.day
````

````python
working_with_date.info()
````

````python
working_with_date[['date','year','month','day']].head(10)
````
